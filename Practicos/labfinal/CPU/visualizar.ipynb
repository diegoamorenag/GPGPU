```python
import cv2

# Load the image in grayscale
image = cv2.imread('../imagenes/1.pgm', cv2.IMREAD_GRAYSCALE)
```

```python
import matplotlib.pyplot as plt

# Display the image
plt.figure(figsize=(12, 12))
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```
